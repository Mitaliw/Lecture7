# Functions

```python
def get_scores_df(report_card_dictionary, division_name):
    
    # convert our dictionary to a dataframe
    report_card_df = pd.DataFrame(report_card_dictionary)
    
    # Add a total score column
    report_card_df["total_Score"] = report_card_df["math_scores"] + report_card_df["chemistry_scores"] + report_card_df["information_technology_scores"] + report_card_df["physics_scores"]
    
    # Add an average score column
    report_card_df["average_score"] = report_card_df["total_Score"]/4
    
    # Add a column called pass fail column
    report_card_df.loc[report_card_df["average_score"] >= 35, "Status"] = "pass"
    report_card_df.loc[report_card_df["average_score"] < 35, "Status"] = "fail"
    
    # Creating a character grade column
    report_card_df.loc[report_card_df["average_score"] >= 90, "character_grade"] = "A"
    report_card_df.loc[(report_card_df["average_score"] >= 70) & (report_card_df["average_score"] < 90), "character_grade"] = "B"
    report_card_df.loc[(report_card_df["average_score"] >= 50) & (report_card_df["average_score"] < 70), "character_grade"] = "C"
    report_card_df.loc[(report_card_df["average_score"] >= 35)& (report_card_df["average_score"] < 50), "character_grade"] = "D"
    report_card_df.loc[report_card_df["average_score"] < 35, "character_grade"] = "F"
    
    # creating a column called division
    report_card_df["division"] = division_name
    
    # Return the dataframe
    return report_card_df
    
```

```python
get_scores_df(div_a_scores_dictionary, "A")
```

## EMAIL FUNCTIONS

```python
def get_emails_df(emails_dictionary, division_name):
    
    # convert our dicitionary into a dataframe
    emails_df = pd.DataFrame(emails_dictionary)
    
    # convert all my emails to lower case
    emails_df["student_email"] = emails_df["student_email"].str.lower()
    
    # create a division column
    emails_df["division"] = division_name
    
    # return the dataframe
    return emails_df
```

```python
get_emails_df(div_a_emails_dictionary, "A")
```

## Combining Functions

```python
def get_scores_and_emails_df(report_card_dictionary, emails_dicitionary, division_name):
    
    # Making my report card dataframe
    report_card_df = get_scores_df(report_card_dictionary, division_name)
    
    # Making my emails dataframe
    emails_df = get_emails_df(emails_dicitionary, division_name)
    
    scores_emails_df = pd.merge(report_card_df, emails_df, how = "left", on = ["student_name", "division"])
    
    return scores_emails_df
```

```python
get_scores_and_emails_df(div_a_scores_dictionary, div_a_emails_dictionary, "A")
```
